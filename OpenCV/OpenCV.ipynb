```python
import cv2
```

```python
# Every image has three channels - R, G, B , a 3d matrix for every pixel.
# The values of these RGB lie between 0 - 255 as represented by 8-bit values of light intensity at every pizel, there for 
# 0 - 255 values

img = cv2.imread('./Images/animal.jpg') # imread is used to read an image
print(img)
```

```python
# Shape of image

img.shape # 492 rows, 980 columns and 3 channel
```

```python
# To see the image

import matplotlib.pyplot as plt

plt.style.use('seaborn') # plt.style.available can be used to see the available styles and plt.style.use to use them
plt.imshow(img) # imshow is used to print the image that have been stored
plt.show()
```

<b> Issue - </b> Whenever open openCV reads the image, it reads image as BGR format and but we enter the image in RGB format. So red gets converted into Blue, Green stays the same and Blue gets converted into Red. This changes our real image.
    
To overcome this, openCV provides us an option to change our color using <code> cvtColor </code> along with <code> COLOR_BGR2RGB </code>

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
# We can also convert to black & white
bg_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(bg_img)
plt.show()
```

```python
# If you want to hide the axis number 
plt.imshow(rgb_img)
plt.axis('off')
plt.show()
```

```python
# Cropping the image

crop_img = rgb_img[30:, 250:750 , :] #[rows, columns, channels]
plt.imshow(crop_img)
plt.axis('off')
plt.show()
```

```python
# We can also see dimensions of cropped image

crop_img.shape
```

Total list that our image store is 462*500*3 i.e. 6,93,000 and every list has 3 RGB values that makes it upto 20,79,000. This is after cropping the value. Original one would have more pixel values. So it is always good thing resize our image.

We can resize in openCV using <code> resize </code>

Therefore we can resize the image to train model to reduce the time and space.

```python
resize_img = cv2.resize(crop_img, [92, 100])
plt.axis('off')
plt.imshow(resize_img)
plt.title('Resized Image')
plt.show()
```

```python
# Size of resized image

resize_img.shape
```

```python
# Percentage reduction 

crop = 462*500
resize = 92*100
print('Percentage decrease in image:', (crop-resize)*100/crop)
```
